```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def scrape_insurance_news():
    url = "https://www.reuters.com/business/finance/"
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    headlines = []
    for item in soup.find_all('h3', limit=5):
        headlines.append(item.get_text())

    df = pd.DataFrame(headlines, columns=['Market Insights'])
    df.to_csv('market_trends.csv', index=False)
    print("Market trends data saved successfully!")


scrape_insurance_news()
```
